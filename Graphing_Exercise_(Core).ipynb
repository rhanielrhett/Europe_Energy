```python
# Import libraries and mount drive
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load the dataset
europe_power_df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Data-file-Europe-Power-Sector-2020.xlsx')
europe_power_df.head()
```

```python
# Check the information of the dataframe
europe_power_df.info()

# Missing values are in columns Share of production and Change on last year
# For the sake of the activity, we will omit for now the data cleaning and will go back to it later on
```

```python
# Pairplot
sns.pairplot(data=europe_power_df)
```

```python
# Bubble plot: year vs. generation based on share in production and variable
sns.set_theme(style='whitegrid')
g = sns.relplot(
    data=europe_power_df,
    x='Year', y='Generation (TWh)',
    hue='Variable', size='Share of production (%)', sizes=(10, 200),
    alpha=0.6
)
plt.xticks(rotation=90)
```

```python
# Bar plot: Variable vs. generation
plt.bar(x='Variable', height='Generation (TWh)', data=europe_power_df)
plt.xticks(rotation=90)
plt.ylim(bottom=0, top=3500)
```

```python
# Bar plot: Area vs. generation
plt.bar(x='Area', height='Generation (TWh)', data=europe_power_df)
plt.xticks(rotation=90)
plt.ylim(bottom=0, top=3500)
```

```python
# Histogram: share of production
plt.hist(x='Share of production (%)', bins=25, data=europe_power_df, cumulative=False)
```

```python
# Histogram: generation
plt.hist(x='Generation (TWh)', bins=25, data=europe_power_df, cumulative=False)
```

```python
# Scatter plot: share in production vs. generation
plt.scatter(x='Share of production (%)', y='Generation (TWh)', marker='x', data=europe_power_df)
```
